```python
!pip install pandas numpy matplotlib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def get_csv_from_local():
    from google.colab import files
    uploaded = files.upload()
    for file in uploaded:
        return pd.read_csv(file)

def get_csv_from_url():
    url = input("Enter the URL of the CSV file: ")
    return pd.read_csv(url)

def get_csv_from_code():
    url = "YOUR_CSV_URL_HERE"
    return pd.read_csv(url)

def save_csv_as_dataframe(csv_data):
    return pd.DataFrame(csv_data)

def print_data(df):
    print("Column Names:")
    print(df.columns.tolist())
    print("\nFirst Two Rows:")
    print(df.head(2))

def convert_columns_to_arrays(df, column1, column2=None):
    if column2:
        return np.array(df[column1]), np.array(df[column2])
    else:
        return np.array(df[column1])

def plot_data(x, y=None):
    plt.figure(figsize=(8, 6))
    if y:
        plt.scatter(x, y)
        plt.xlabel('X-axis')
        plt.ylabel('Y-axis')
        plt.title('Scatter Plot')
    else:
        plt.plot(x)
        plt.xlabel('Index')
        plt.ylabel('Values')
        plt.title('Line Graph')
    plt.show()

def main():
    print("1. Get CSV from local")
    print("2. Get CSV from URL")
    print("3. Get CSV from code")
    choice = input("Enter your choice (1-3): ")

    if choice == "1":
        csv_data = get_csv_from_local()
    elif choice == "2":
        csv_data = get_csv_from_url()
    elif choice == "3":
        csv_data = get_csv_from_code()
    else:
        print("Invalid choice. Exiting...")
        return

    df = save_csv_as_dataframe(csv_data)
    print_data(df)

    column_names = df.columns.tolist()
    print("Column names list:", column_names)

    print("Select column(s) for conversion:")
    print("Available columns:", column_names)
    column1 = input("Enter the name of column 1: ")
    column2 = input("Enter the name of column 2 (optional, leave blank for single column): ")

    x, y = convert_columns_to_arrays(df, column1, column2)
    plot_data(x, y)

if __name__ == "__main__":
    main()
```
