```python
from fractions import Fraction
import math

def solve_proportion(a, b, c):
    """Solves a proportion of the form a/b = c/x"""
    x = (b * c) / a
    return x

def solve_equation(equation):
    """Solves an equation for x"""
    x = eval(equation)
    return x

def factor_square_root(number):
    """Factors a square root of a number"""
    factors = []
    for i in range(2, int(math.sqrt(number)) + 1):
        while number % (i * i) == 0:
            factors.append(i)
            number //= (i * i)
    if number > 1:
        factors.append(number)
    return factors

def decimal_to_fraction(decimal):
    """Converts a decimal to a fraction"""
    fraction = Fraction(decimal).limit_denominator()
    return fraction

def decimal_to_percent(decimal):
    """Converts a decimal to a percentage"""
    percent = decimal * 100
    return percent

def fraction_to_decimal(fraction):
    """Converts a fraction to a decimal"""
    decimal = float(fraction)
    return decimal

def fraction_to_percent(fraction):
    """Converts a fraction to a percentage"""
    percent = fraction * 100
    return percent

def percent_to_decimal(percent):
    """Converts a percentage to a decimal"""
    decimal = percent / 100
    return decimal

def percent_to_fraction(percent):
    """Converts a percentage to a fraction"""
    fraction = Fraction(percent/100).limit_denominator()
    return fraction

# Main program
while True:
    print("----- Calculator Menu -----")
    print("1. Solve Proportions")
    print("2. Solve Equations for x")
    print("3. Factor Square Roots")
    print("4. Convert Decimals to Fractions and Percents")
    print("5. Convert Fractions to Decimals and Percents")
    print("6. Convert Percents to Decimals and Fractions")
    print("0. Exit")

    choice = input("Enter your choice (0-6): ")

    if choice == "0":
        break

    if choice == "1":
        a = float(input("Enter the value of a: "))
        b = float(input("Enter the value of b: "))
        c = float(input("Enter the value of c: "))
        result = solve_proportion(a, b, c)
        print(f"The value of x is: {result}")

    elif choice == "2":
        equation = input("Enter the equation in terms of x: ")
        result = solve_equation(equation)
        print(f"The value of x is: {result}")

    elif choice == "3":
        number = int(input("Enter the number to factor the square root: "))
        factors = factor_square_root(number)
        print(f"The factors of the square root of {number} are: {factors}")

    elif choice == "4":
        decimal = float(input("Enter the decimal number: "))
        fraction = decimal_to_fraction(decimal)
        percent = decimal_to_percent(decimal)
        print(f"The fraction is: {fraction}")
        print(f"The percent is: {percent}%")

    elif choice == "5":
        numerator = int(input("Enter the numerator of the fraction: "))
        denominator = int(input("Enter the denominator of the fraction: "))
        fraction = Fraction(numerator, denominator)
        decimal = fraction_to_decimal(fraction)
        percent = fraction_to_percent(fraction)
        print(f"The decimal is: {decimal}")
        print(f"The percent is: {percent}%")

    elif choice == "6":
        percent = float(input("Enter the percent: "))
        decimal = percent_to_decimal(percent)
        fraction = percent_to_fraction(percent)
        print(f"The decimal is: {decimal}")
        print(f"The fraction is: {fraction}")

    print()  # Empty line for readability

print("Exiting the calculator. Goodbye!")
```
