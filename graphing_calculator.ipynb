```python
import numpy as np
import matplotlib.pyplot as plt

def plot_function(expression, x_range=(-10, 10), y_range=(-10, 10)):
    x = np.linspace(x_range[0], x_range[1], 1000)
    y = eval(expression, {'x': x})
    
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.xlim(x_range)
    plt.ylim(y_range)
    plt.grid(True)
    plt.show()

def create_table(expression, x_range=(-10, 10)):
    x = np.linspace(x_range[0], x_range[1], 100)
    y = eval(expression, {'x': x})
    
    table = list(zip(x, y))
    print(" x     |     y")
    print("-------|-------")
    for row in table:
        print("{:6.2f} | {:6.2f}".format(*row))

def shade_above(expression, x_range=(-10, 10), y_range=(-10, 10)):
    x = np.linspace(x_range[0], x_range[1], 1000)
    y = eval(expression, {'x': x})
    
    plt.fill_between(x, y, y_range[1], color='gray')
    plt.plot(x, y, color='black')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.xlim(x_range)
    plt.ylim(y_range)
    plt.grid(True)
    plt.show()

def solve_system_equations(eq1, eq2, x_range=(-10, 10), y_range=(-10, 10)):
    x = np.linspace(x_range[0], x_range[1], 1000)
    y1 = eval(eq1, {'x': x})
    y2 = eval(eq2, {'x': x})
    
    plt.plot(x, y1, label='Equation 1')
    plt.plot(x, y2, label='Equation 2')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.xlim(x_range)
    plt.ylim(y_range)
    plt.grid(True)
    plt.legend()
    plt.show()

def zoom_graph(expression, x_range=(-10, 10), y_range=(-10, 10)):
    x = np.linspace(x_range[0], x_range[1], 1000)
    y = eval(expression, {'x': x})
    
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.gca().set_aspect('equal')
    plt.xlim(x_range)
    plt.ylim(y_range)
    plt.show()

def solve_quadratic_equation(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant < 0:
        return "No real solutions"
    elif discriminant == 0:
        x = -b / (2*a)
        return "One real solution: x = {}".format(x)
    else:
        x1 = (-b + np.sqrt(discriminant)) / (2*a)
        x2 = (-b - np.sqrt(discriminant)) / (2*a)
        return "Two real solutions: x1 = {}, x2 = {}".format(x1, x2)

# Example usage
plot_function('x**2')
create_table('x**2')
shade_above('x**2')
solve_system_equations('x', '2*x')
zoom_graph('x**2', x_range=(-5, 5), y_range=(-5, 5))
print(solve_quadratic_equation(1, -4, 3))
```
